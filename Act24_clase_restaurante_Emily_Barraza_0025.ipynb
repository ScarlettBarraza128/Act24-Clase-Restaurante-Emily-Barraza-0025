# Clase restaurante

* Atributos: nombre, ubicacion, capacidad, tipoCocina
* Funciones: recibirReserva(), prepararPedido(), facturar()


![texto del vínculo](https://raw.githubusercontent.com/ScarlettBarraza128/Act24-Clase-Restaurante-Emily-Barraza-0025/refs/heads/main/InfoClaseRestauranteEmily0025.png)

```python
class Restaurante:
    # 1. Metodo __init__ para inicializar atributos (constructor)
    def __init__(self, nombre, ubicacion, capacidad, tipoCocina):
        # Atributos encapsulados (datos únicos)
        self.nombre = nombre
        self.ubicacion = ubicacion
        self.capacidad = capacidad
        self.tipoCocina = tipoCocina
        self.abierto = False

    # 2. Comportamientos (métodos)
    def recibirReserva(self):
        print(f"📅 Reserva recibida en '{self.nombre}'. ¡Preparando su mesa! 🪑")

    def prepararPedido(self):
        print(f"👨‍🍳 Cocinando especialidad {self.tipoCocina} en {self.nombre}... 🥘🔥")

    def facturar(self):
        print(f"🧾 Factura generada para la sucursal: {self.ubicacion}. ¡Gracias por su compra! 💰")

# 3. Instanciar múltiples objetos con características únicas
restaurante1 = Restaurante("La Trattoria", "Av. Italia 500", 40, "Italiana 🍝")
restaurante2 = Restaurante("Dragon Roll", "Calle Oriente 25", 25, "Japonesa 🍣")

# Llamadas a las funciones para cada objeto restaurante 1 y restaurante 2
print("📍 Restaurante 1:")
restaurante1.recibirReserva()
restaurante1.prepararPedido()
restaurante1.facturar()

print("\n📍 Restaurante 2:")
restaurante2.prepararPedido()
restaurante2.recibirReserva()
restaurante2.facturar()

print("\n✨ Creador de software por Emily Barraza NL = 7 ✨")
print("🏁 programa terminado")
```
